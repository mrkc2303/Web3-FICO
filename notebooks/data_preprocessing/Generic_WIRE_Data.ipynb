```python
# Step 1: Create the new format for df_malicious
df_malicious_formatted = df_malicious.copy()
df_malicious_formatted['Name'] = 'NA'  # Add Name column with 'NA'
df_malicious_formatted['isSafe'] = False  # Add isSafe column as False
df_malicious_formatted['Tags'] = 'NA'  # Add Tags column with 'NA'
df_malicious_formatted = df_malicious_formatted.rename(columns={'malicious_wallet_address': 'Address'})

# Step 2: Format wallets_df to match the desired output
wallets_df_formatted = wallets_df.copy()
wallets_df_formatted['isSafe'] = wallets_df_formatted['Label'].apply(lambda x: True if x == 'Legit' else False)
wallets_df_formatted = wallets_df_formatted[['Address', 'Name', 'isSafe', 'Tags']]

# Step 3: Concatenate the two DataFrames
combined_df = pd.concat([wallets_df_formatted, df_malicious_formatted], ignore_index=True)

# Step 4: Shuffle the combined DataFrame to make it random
combined_df = combined_df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
import pandas as pd
import requests
import time

# Load combined_df (replace with the actual CSV or DataFrame)
combined_df = pd.read_csv('combined_wallets.csv')

# Initialize new columns
combined_df['transactions'] = None
combined_df['balance'] = None
combined_df['noOfTrx'] = 0  # Initialize the new column with 0

# Etherscan API base URL and API key
api_url_trx = "https://api.etherscan.io/v2/api"
api_url_balance = "https://api.etherscan.io/v2/api"
api_key = "YourApiKeyToken"  # Replace with your actual API key

# Set testing mode
testing = False  # Set to False for full run

# Loop through the addresses in combined_df
addresses = combined_df['Address'].tolist()

# Initialize counters
result_transactions = []
result_balances = []

# Chunk addresses for balance API (5 per request)
chunks = [addresses[i:i + 5] for i in range(0, len(addresses), 5)]

# Loop through chunks
for i, chunk in enumerate(chunks):
    if testing and i > 0:  # For testing, run the loop only once
        break

    # Call balance API for the chunk
    balance_params = {
        "chainid": 1,
        "module": "account",
        "action": "balancemulti",
        "address": ",".join(chunk),
        "tag": "latest",
        "apikey": "7A32HIJRFBSUJQK8S6WD3E2XCDS6XBW9JZ"
    }
    balance_response = requests.get(api_url_balance, params=balance_params)
    
    if balance_response.status_code == 200:
        balance_data = balance_response.json()
        if balance_data['status'] == "1":
            for item in balance_data['result']:
                balance_address = item['account']
                balance_value = item['balance']
                # Add balance to result_balances
                result_balances.append({"Address": balance_address, "Balance": balance_value})
        else:
            print(f"Balance API Error: {balance_data['message']}")
    else:
        print(f"API Call Failed: {balance_response.status_code}")

    # Call transaction API for each address in the chunk
    for address in chunk:
        trx_params = {
            "chainid": 1,
            "module": "account",
            "action": "txlist",
            "address": address,
            "startblock": 0,
            "endblock": 99999999,
            "sort": "asc",
            "apikey": "7A32HIJRFBSUJQK8S6WD3E2XCDS6XBW9JZ"
        }
        trx_response = requests.get(api_url_trx, params=trx_params)
        
        if trx_response.status_code == 200:
            trx_data = trx_response.json()
            if trx_data['status'] == "1":
                # Append transactions to the result
                result_transactions.append({
                    "Address": address, 
                    "Transactions": trx_data['result'],
                    "noOfTrx": len(trx_data['result'])  # Add this to count transactions
                })
                print(f"No of Trx Found: {len(trx_data['result'])}")

            else:
                print(f"Transaction API Error: {trx_data['message']}")
        else:
            print(f"API Call Failed: {trx_response.status_code}")
        
        # Wait 1 second between API calls
        time.sleep(2)

# Update combined_df with results
transactions_df = pd.DataFrame(result_transactions)
balances_df = pd.DataFrame(result_balances)

# Merge transactions and balances into combined_df
combined_df = combined_df.merge(transactions_df, on='Address', how='left')
combined_df = combined_df.merge(balances_df, on='Address', how='left')

# Ensure 'noOfTrx' is properly initialized
if 'noOfTrx' not in combined_df.columns:
    combined_df['noOfTrx'] = 0  # Initialize with 0 for missing data
else:
    combined_df['noOfTrx'] = combined_df['noOfTrx'].fillna(0).astype(int)

# Save to a new CSV file
output_file = 'combined_wallets_with_transactions_and_balances.csv'
combined_df.to_csv(output_file, index=False)

print(f"Data saved to {output_file}")
```
